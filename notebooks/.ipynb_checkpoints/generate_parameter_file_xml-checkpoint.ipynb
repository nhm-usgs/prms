### Generate a basic XML file for the control parameters

```python
import xml.dom.minidom as minidom
import xml.etree.ElementTree as xmlET

ctl_xml = xmlET.Element('parameters')

#     <parameter name="ssstor_init_frac">
#         <type>F</type>
#         <units>decimal fraction</units>
#         <desc>Initial fraction of available water in the gravity plus preferential-flow reservoirs</desc>
#         <help>Initial fraction of available water in the gravity plus preferential-flow reservoirs (fraction of sat_threshold) for each HRU</help>
#         <minimum>0.0</minimum>
#         <maximum>1.0</maximum>
#         <default>0.0</default>
#         <modules>
#             <module>soilzone</module>
#         </modules>
#     </parameter>


for index, row in df.iterrows():
    print(row)
    ctl_entry = xmlET.SubElement(ctl_xml, 'parameter')
    ctl_entry.set('name', row['parameter'])
    
    # TODO: type goes here
    xmlET.SubElement(ctl_entry, 'default').text = row['default_value']
    xmlET.SubElement(ctl_entry, 'units').text = row['units']
    xmlET.SubElement(ctl_entry, 'desc').text = row['description']
    xmlET.SubElement(ctl_entry, 'help').text = row['help']
    xmlET.SubElement(ctl_entry, 'minimum').text = row['min_value']
    xmlET.SubElement(ctl_entry, 'maximum').text = row['max_value']
    
    xmlET.SubElement(ctl_entry, 'type').text = "1"
    xmlET.SubElement(ctl_entry, 'numvals').text = "1"
    xmlET.SubElement(ctl_entry, 'desc').text = "empty"
    
    print(row['ctl_param'], row['default'])
```

### Write XML file

```python
xmlstr = minidom.parseString(xmlET.tostring(ctl_xml)).toprettyxml(indent='    ')
with open('/Users/pnorton/tmp/control.xml', 'w') as ff:
    ff.write(xmlstr)
    
# print(xmlstr)
```
